### Exploratory Data Analysis

```python
df["less than 10 min ratio"] = df['mean_Less than 10 min']/df['Estimate!!Aggregate travel time to work (in min):']
df["more than 60 min ratio"] = df['mean_60 or more min']/df['Estimate!!Aggregate travel time to work (in min):']
```

```python
sns.set();
df.hist(figsize=(15,10));
```

```python
#visualization of Date for less than 10 mins

plt.figure(figsize=(10,10))

# x-axis
plt.xlabel('Aggregated Mean ratio', fontsize = 11)
# y-axis
plt.ylabel('Counts', fontsize = 11)
# Title
plt.title('Commute time comparision')
# Legend 

# Plot Line chart
plt.plot(df["less than 10 min ratio"],color='blue')
plt.plot(df["more than 60 min ratio"], color ='pink')

# Display plot
plt.show()
```

```python
#making heatmap for two columns

cols = ["less than 10 min ratio", "more than 60 min ratio"]
df_corr = df[cols].corr()

# Generate the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, cmap="YlGnBu", annot=True)
plt.title("Correlation Heatmap between {} and {}".format(cols[0], cols[1]))
plt.show()
```
